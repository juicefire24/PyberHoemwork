```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
# import seaborn as sns; sns.set()
# sns.set_color_codes()

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)

# Combine the data into a single dataset
pyber_df = pd.merge(ride_data, city_data, on="city", how="left")

# Display the data table for preview
pyber_df.head()

# # city_data.mean()
# ride_data.mean()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the data frames for the city type(urban_rural,suburban) and then group the data frames by city
# Desired colurs that need to be provided for lables
# colors = ['gold', 'lightskyblue', 'coral']
# labels = ['urban', 'rural', 'suburban']
# Obtain the x and y coordinates for each of the three city types urban,rural, suburban

urban_city_type = pyber_df[pyber_df["type"] == "Urban"].groupby([pyber_df["city"]])
rural_city_type = pyber_df[pyber_df["type"] == "Rural"].groupby([pyber_df["city"]])
suburban_city_type = pyber_df[pyber_df["type"] == "Suburban"].groupby([pyber_df["city"]])

# Urban calculations of ride count (x), fare average(y), driver count average(s as in size)
x_urban = urban_city_type["ride_id"].count()
y_urban = urban_city_type["fare"].mean()
s_urban = urban_city_type["driver_count"].mean()

# Rural calculations of ride count (x), fare average(y), driver count average(s as in size)
x_rural = rural_city_type["ride_id"].count()
y_rural = rural_city_type["fare"].mean()
s_rural = rural_city_type["driver_count"].mean()

# Suburban calculations of ride count (x), fare average(y), driver count average(s as in size)
x_suburban = suburban_city_type["ride_id"].count()
y_suburban = suburban_city_type["fare"].mean()
s_suburban = suburban_city_type["driver_count"].mean()

# Build the scatter plots for each city types
plt.scatter(x_urban, y_urban, label = "Urban", s=s_urban * 9, color=["coral"], edgecolor="black", alpha = 0.5, marker="o")
plt.scatter(x_rural, y_rural, label = "Rural", s=s_rural * 9, color=["gold"], edgecolor="black", alpha = 0.7, marker="o")
plt.scatter(x_suburban, y_suburban, label = "Suburban", s=s_suburban  * 9, color=["lightskyblue"], edgecolor="black", alpha = 0.8, marker="o")
plt.grid()

# Incorporate the other graph properties xlabel,ylabel,title, and note
plt.xlabel("Total Number of Rides (Per City)")
plt.ylabel("Average Fare ($)", fontsize = 14)
plt.title("Pyber Ride Sharing Data (2016)")
plt.text(42,35,"Note: \nCircle size correlates with driver count per city.", fontsize = 12)

# Create a legend
legend = plt.legend(title="City Types",loc="best",fontsize=9)
for l in legend.legendHandles:
    l.set_sizes([30.0])
    
# Save Figure
plt.savefig("PyPies.png")
    
# Show plot
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
# Creates the pie chart based upon the values above
# Automatically finds the percentages of each part of the pie chart

type_groupped = pyber_df.groupby(['type'])
fare_sum = type_groupped['fare'].sum()

# Build Pie Chart
labels = ["Rural","Suburban","Urban"]
# The values of each section of the pie chart
sizes = [200, 90, 70,]
colors = ["gold", "lightskyblue", "lightcoral"]
explode = (0, 0, 0.1)

plt.pie(fare_sum, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=170)
plt.title("% of Total Fares by City Type")

# Tells matplotlib that we want a pie chart with equal axes
plt.axis("equal")

# Save an image of our chart and print the final product to the screen
plt.savefig("PyPies2.png")

# Show Figure
plt.show()
```
